```
import pandas as pd
from apyori import apriori
```

```
store_data = pd.read_csv("/content/drive/My Drive/TRAIN-ARULES.csv")
print(store_data.head())
```

.


.

**Train Data...**

```

product_data = store_data.groupby('order_id')['product_name']

records = []
for order, pName in product_data:
    records.append(pName)

#association_rules = apriori(records, min_support=0.0045, min_confidence=0.2)
association_rules = apriori(records, min_support=0.0045, use_colnames=True)
association_results = list(association_rules)
```

```

print("Association results begins....")
print("*********************************************************************************************************************")
print("\n")

Count = 0

for result in association_results:
    items = [x for x in result.items]
    print("Items   ="+str(items))
    print("Support ="+str(result.support))
    for OrderedStatistic in result.ordered_statistics:
        items_base = [x for x in OrderedStatistic.items_base]
        items_add = [x for x in OrderedStatistic.items_add]
        print ("  Rule       = " + str(items_base) + " --> " + str(items_add))
        print ("  Confidence = " + str(OrderedStatistic.confidence))
        print ("  Lift       = " + str(OrderedStatistic.lift))
        break
    print("----------------------------------------")
print("\n")
print("*********************************************************************************************************************")
print("Association results ends......")
```

```


print("Total Association Results = "+str(len(association_results)))

```

.

**Test Data.....**

```
test_data = pd.read_csv('/content/drive/My Drive/testarules.csv').dropna(axis='columns')

print("Test results begins....")
print("*********************************************************************************************************************")
print("\n")
print("Test data    = "+str(test_data.values.tolist()[0]))
testdata = test_data.values.tolist()[0]

for result in association_results:
    items = [x for x in result.items]
  #  print("Count "+str(Count)+" Items :"+str(items) + "Test "+ str(testdata))
    if (all(x in items for x in testdata)):
         Count = Count + 1
         print("Test Support = "+str(result.support))
         print("----------------------------------------")
         for OrderedStatistic in result.ordered_statistics:
             items_base = [x for x in OrderedStatistic.items_base]
             items_add = [x for x in OrderedStatistic.items_add]
             print ("  Rule       = " + str(items_base) + " --> " + str(items_add))
             print ("  Confidence = " + str(OrderedStatistic.confidence))
             print ("  Lift       = " + str(OrderedStatistic.lift))
             print ("  -----------------------------")
print("\n")
print("*********************************************************************************************************************")
print("Test results ends......")

```
